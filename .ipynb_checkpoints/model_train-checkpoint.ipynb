```python
def my_loss(y_true,y_pred):
    return K.sqrt(K.mean(K.square(y_true[...,0:3]-y_pred[...,0:3])))
    

def net(input_size=(IMAGE_SHAPE[0],IMAGE_SHAPE[0],3)):
    x_left=Input(shape=(input_size))
    x_right=Input(shape=(input_size))
    
    x3=Conv2D(50,(3,3),padding='same',activation='relu')(x_left)
    x4=Conv2D(10,(4,4),padding='same',activation='relu')(x_left)
    x5=Conv2D(5,(5,5),padding='same',activation='relu')(x_left)
    x=concatenate([x3,x4,x5])
    
    x3=Conv2D(50,(3,3),padding='same',activation='relu')(x)
    x4=Conv2D(10,(4,4),padding='same',activation='relu')(x)
    x5=Conv2D(5,(5,5),padding='same',activation='relu')(x)
    x_l=concatenate([x3,x4,x5])
    
    x3=Conv2D(50,(3,3),padding='same',activation='relu')(x_right)
    x4=Conv2D(10,(4,4),padding='same',activation='relu')(x_right)
    x5=Conv2D(5,(5,5),padding='same',activation='relu')(x_right)
    x=concatenate([x3,x4,x5])
    
    x3=Conv2D(50,(3,3),padding='same',activation='relu')(x)
    x4=Conv2D(10,(4,4),padding='same',activation='relu')(x)
    x5=Conv2D(5,(5,5),padding='same',activation='relu')(x)
    x_r=concatenate([x3,x4,x5])
    
    x=concatenate([x_l,x_r])
    
    x3=Conv2D(50,(3,3),padding='same',activation='relu')(x)
    x4=Conv2D(10,(4,4),padding='same',activation='relu')(x)
    x5=Conv2D(5,(5,5),padding='same',activation='relu')(x)
    x=concatenate([x3,x4,x5])
    
#     x3=Conv2D(50,(3,3),padding='same',activation='relu')(x)
#     x4=Conv2D(10,(4,4),padding='same',activation='relu')(x)
#     x5=Conv2D(5,(5,5),padding='same',activation='relu')(x)
#     x=concatenate([x3,x4,x5])
    
#     x3=Conv2D(50,(3,3),padding='same',activation='relu')(x)
#     x4=Conv2D(10,(4,4),padding='same',activation='relu')(x)
#     x5=Conv2D(5,(5,5),padding='same',activation='relu')(x)
#     x=concatenate([x3,x4,x5])
    
    output=Conv2D(3,(3,3),padding='same',activation='relu')(x)
    model=Model(inputs=[x_left,x_right],outputs=output)
    model.compile(optimizer=SGD(),loss=my_loss)
    
    return model
```

```python
model=net()
```

```python
#
from tqdm import *
```

```python
#
EPOCHs=5000
BACRH_SIZE=2
m=x_left.shape[0]
loss_history=[]
for epoch in range(EPOCHs):
    t=tqdm(range(0,x_left.shape[0],BACRH_SIZE),mininterval=0)
    loss=[]
    for idx in t:
        batch_left=x_left[idx:min(idx+BACRH_SIZE,m)]
        batch_right=x_right[idx:min(idx+BACRH_SIZE,m)]
        batch_y=y[idx:min(idx+BACRH_SIZE,m)]
        loss.append(model.train_on_batch([batch_left,batch_right],batch_y))
        t.set_description('Epoch{} | Batch{} of {}.Loss {:f}'.format(epoch+1,idx,m,np.mean(loss)))
    loss_history.append(np.mean(loss))
```
